```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('medical_examination.csv')
df
```

```python
df['overweight'] = (df['weight'] * 0.45359237) / ((df['height'] / 100) ** 2)
df['overweight'] = df['overweight'].apply(lambda x: 0 if x <= 25 else 1)
df
```

```python
df['cholesterol'] = df['cholesterol'].apply(lambda x: 0 if x == 1 else 1)
df['gluc'] = df['gluc'].apply(lambda x: 0 if x == 1 else 1)
df
```

```python
prep_df = df[['id', 'cholesterol', 'gluc', 'smoke', 'alco', 'active', 'overweight']]
long = pd.melt(prep_df, id_vars = 'id', var_name='variable', value_name='value')
long
```

```python
cardio = df[['id', 'cardio']]
long = long.merge(cardio, on='id', how='left')
long
```

```python
sns.catplot(kind = 'count', data = long, x = 'variable', hue = 'value', col = 'cardio')
```

```python
df = df[df['height'] >= df['height'].quantile(0.025)] # fix, along with other fixed issues
df = df[df['height'] <= df['height'].quantile(0.975)]
df = df[df['weight'] >= df['weight'].quantile(0.025)]
df = df[df['weight'] <= df['weight'].quantile(0.975)] # height and weight should be in between the 2.5th and 97.5th percentile of the data set.
```

```python
no_nan_corr = corr.iloc[:-1, :-1].dropna(axis=0, how='all').dropna(axis=1, how='all')
no_nan_corr # overweight has nan values due to constant values in original df
```
